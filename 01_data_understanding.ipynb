## Sales Trend Analysis

In this section, we analyze sales performance over time by extracting year and month information from the order date. This enables us to identify growth trends, seasonality, and year-over-year performance.

```python
# Extract the order year for yearly trend analysis
df['Order Year'] = df['Order Date'].dt.year

# Extract the month name for monthly reporting
df['Order Month Name'] = df['Order Date'].dt.month_name()

# Extract the month number to maintain chronological month sorting
df['Order Month Num'] = df['Order Date'].dt.month
```

## Yearly Sales and Profit Analysis

This analysis examines yearly sales and profit trends to evaluate business growth over time and identify changes in overall performance.

```python
# Calculate yearly sales and profit to analyze business performance over time


yearly_performance  = (
    df.groupby('Order Year')[["Sales", "Profit"]]
    .sum()
    .reset_index()
)

yearly_performance
```

```python
# Analyze average profit by discount level

discount_analysis = (
    df.groupby('Discount')[["Sales","Profit"]]
    .mean()
    .reset_index()
    .sort_values("Discount")
)
discount_analysis
```

```python
# Analyze sales and profit by product category
category_performance = (
    df.groupby('Category')[["Sales", "Profit"]]
    .sum()
    .sort_values("Sales", ascending=False)
)

category_performance
```

## Furniture Sub-Category Analysis

This analysis drills down into the Furniture category to identify which sub-categories contribute positively or negatively to overall profitability.

```python
# Analyze sales and profit for Furniture sub-categories

furniture_performance = (
    df.groupby('Sub-Category')[['Sales', 'Profit']]
    .sum()
    .sort_values('Profit', ascending=False)
)

furniture_performance
```

## Furniture Profit by Region

This analysis evaluates the profitability of Furniture across different regions to determine whether losses are concentrated in specific geographic areas.

```python
# Analyze sales and profit across regions for the Furniture category to identify geographic areas contributing to low profitability
furniture_analysis = (
    df[df['Category'] == 'Furniture']
    .groupby('Region')[['Sales', 'Profit']]
    .sum()
    .sort_values('Profit', ascending=False)
)

furniture_analysis
```

## Discount Analysis for Furniture in the Central Region

This analysis investigates whether high discounts are contributing to the negative profitability of Furniture products in the Central region.

```python
# Filter the dataset to include only Furniture sales in the Central region

central_furniture_sales = df[(df['Category'] == 'Furniture') & (df['Region'] == 'Central')]

# Sort the results by profit to identify the most and least profitable discount levels
central_discount_analysis = (
    central_furniture_sales.groupby('Discount')[['Sales', 'Profit']]
    .sum()
    .sort_values('Profit', ascending=False)
)

central_discount_analysis
```

## Product-Level Analysis for Furniture in the Central Region

This analysis identifies the least profitable Furniture products in the Central region to determine whether losses are concentrated in a small number of products or distributed across the entire category.

```python
# Filter the dataset to include only Furniture sales in the Central region
central_furniture_sales = df[(df['Category'] == 'Furniture') & (df['Region'] == 'Central')]

# Analyze sales and profit by product to identify the least profitable products
central_furniture_product_analysis = (
    central_furniture_sales.groupby('Product Name')[['Sales', 'Profit']]
    .sum()
    .sort_values('Profit', ascending=True)
)

central_furniture_product_analysis
```

## Category Profit Margin Analysis

This analysis evaluates the profit margin for each product category to measure profitability relative to sales rather than relying on profit alone.

```python
# Calculate total sales and profit for each product category
total_sales_profit_category = (
    df.groupby('Category')[['Sales', 'Profit']]
    .sum()
    .sort_values('Profit', ascending=False)
    .reset_index()
)
total_sales_profit_category
```

```python
# Calculate the profit margin for each product category
total_sales_profit_category['Profit Margin'] = (
    (total_sales_profit_category['Profit'] / total_sales_profit_category['Sales']) * 100

)
total_sales_profit_category[['Category', 'Profit Margin']]
```

```python
# Sort categories by profit margin to compare overall profitability
total_sales_profit_category.sort_values('Profit Margin', ascending=False)
```

## Shipping Mode Analysis for Furniture in the Central Region

This analysis investigates whether specific shipping modes are associated with lower profitability for Furniture sales in the Central region.

```python
# Filter the dataset to include only Furniture sales in the Central region
furniture_central_region = df[(df['Category'] == 'Furniture') & (df['Region'] == 'Central')]
```

```python
# Analyze sales and profit by shipping mode
furniture_central_region_shipping_analysis = (
    furniture_central_region.groupby('Ship Mode')[['Sales', 'Profit']]
    .sum()
    .reset_index()
)
furniture_central_region_shipping_analysis
```

```python
# Sort shipping modes by profit to identify the least profitable delivery methods
furniture_central_region_shipping_analysis.sort_values('Profit', ascending=False)
```

## Shipping Mode Performance Analysis

This analysis compares shipping modes using both total and per-order metrics to determine whether profitability differences are driven by shipping performance or simply by order volume.

```python
# Analyze order volume, sales, and profit by shipping mode
furniture_central_region_order_analysis = (
    furniture_central_region.groupby('Ship Mode').agg(
        Total_Sales=('Sales', 'sum'),
        Total_Profit=('Profit', 'sum'),
        Order_Count=('Order ID', 'nunique'),
    )
)

furniture_central_region_order_analysis
```

```python
# Calculate average sales and profit per order for each shipping mode
furniture_central_region_order_analysis['average_sales_per_order'] = furniture_central_region_order_analysis['Total_Sales'] / furniture_central_region_order_analysis['Order_Count']
furniture_central_region_order_analysis['average_profit_per_order'] = furniture_central_region_order_analysis['Total_Profit'] / furniture_central_region_order_analysis['Order_Count'] 

furniture_central_region_order_analysis.sort_values('average_profit_per_order', ascending=False)
```
